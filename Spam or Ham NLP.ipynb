```python
import nltk
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
import pandas
df_sms = pd.read_csv('spam.csv',encoding='latin-1')
df_sms.head()
```

```python
## Dropping the unwanted columns Unnamed:2, Unnamed: 3 and Unnamed:4
df_sms = df_sms.drop(["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1)
```

```python
## Rename the Columns for better understanding
df_sms = df_sms.rename(columns={"v1":"label", "v2":"sms"})

df_sms.head()
```

```python
## Check for null values
df_sms.isnull().sum()
```

```python
df_sms.loc[:,'label'] = df_sms.label.map({'ham':0, 'spam':1})
df_sms.head()
```

```python
df_sms.label.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_sms['sms'],df_sms['label'],test_size=0.20,random_state=1)
```

```python
# Instantiate the CountVectorizer method
count_vector = CountVectorizer()

# Fit the training data and then return the matrix
training_data = count_vector.fit_transform(X_train)

# Transform testing data and return the matrix. 
testing_data = count_vector.transform(X_test)
```

```python
df = pd.DataFrame(training_data.toarray())
df
```

```python
y_train
```

```python
naive_bayes = MultinomialNB()
naive_bayes.fit(training_data,y_train)
```

```python
predictions = naive_bayes.predict(testing_data)
```

```python
# Compare result
result=pd.DataFrame({"Actual":y_test,"Predicted":predictions})
result
```

```python
print('Accuracy score: {}'.format(accuracy_score(y_test, predictions)))
print('Precision score: {}'.format(precision_score(y_test, predictions)))
print('Recall score: {}'.format(recall_score(y_test, predictions)))
print('F1 score: {}'.format(f1_score(y_test, predictions)))
```
